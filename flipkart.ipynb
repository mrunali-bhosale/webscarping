```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from urllib.request import urlopen
```

```python
url = "https://www.flipkart.com/womens-clothing/ethnic-wear/kurtas-kurtis/pr?sid=2oq,c1r,3pj,ua6&otracker=categorytree&otracker=nmenu_sub_Women_0_Kurtas%20%26%20Kurtis"
html = urlopen("https://www.flipkart.com/womens-clothing/ethnic-wear/kurtas-kurtis/pr?sid=2oq,c1r,3pj,ua6&otracker=categorytree&otracker=nmenu_sub_Women_0_Kurtas%20%26%20Kurtis")
```

```python
#print(soup.prettify())
```

```python
pages = []
products = {}
product_count = 0

for i in range (1,2):
    URL = url + '&page='+ str(i)
    pages.append(URL)  

for item in pages :
    page = requests.get(item)
    soup = BeautifulSoup(page.text,'html.parser')
    rows = soup.find_all('div',{'class':'IIdQZO _1R0K0g _1SSAGr'})
    for row in rows:
        Brand  = row.find('div',{'class':'_2B_pmu'}).text
    
        Category_tag = row.find('a',{'class':'_2mylT6'})
        Category = Category_tag.text if Category_tag else "N/A"
        
        DisctPrice_tag = row.find('div',{'class':'_1vC4OE'})
        DisctPrice =  DisctPrice_tag.text if DisctPrice_tag else "N/A"
        
        MRP_tag = row.find('div',{'class':'_3auQ3N'})
        MRP = MRP_tag.text if MRP_tag else "N/A"
    
        Discpercent = row.find('div',{'class':'VGWI6T'})
        Discpercent = Discpercent_tag.text if Discpercent_tag else "N/A"
        
        product_count += 1
        products[product_count] = [Brand, Category,DisctPrice, MRP,Discpercent]
        
        
        print('\nBRAND :',Brand,'\nCATEGORY :',Category,'\nMRP:',MRP,'\nDISTPRICE :',DisctPrice,'\nDIST%',Discpercent)
        
print("Product Count :",product_count)
products_df = pd.DataFrame.from_dict(products,orient = 'index',columns=['BRAND','CATEGORY','DISTPRICE','MRP','DIST%'])

```
